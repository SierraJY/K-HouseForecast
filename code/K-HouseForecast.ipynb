```python
import matplotlib.pyplot as plt
import seaborn as sns



# 지역별 가격 변화 시각화 시, 서로 구별되는 색상을 사용하여 각 지역을 명확하게 구분
plt.figure(figsize=(15, 8))

# 색상 팔레트 생성 (데이터 열 수만큼의 색상)
palette = sns.color_palette("hsv", len(df2.columns) - 1)

# 각 지역별 데이터 시각화
for idx, column in enumerate(df2.columns[1:]):  # 첫 번째 열(전국 소계)는 제외하고 시각화
    plt.plot(df2[column], label=column, color=palette[idx])

# 그래프 설정
plt.title('지역별 아파트 매매 가격 변화')
plt.xlabel('시간(데이터 인덱스)')
plt.ylabel('가격(단위: 만원)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # 범례를 그래프 밖에 위치시킴
plt.grid(True)

plt.xticks(rotation=90)

# 그래프 표시
plt.tight_layout()  # 레이아웃 조정
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd



# 데이터 전처리: 지역별 데이터를 하나의 열로 변환
df_melted = df2.melt(id_vars=[], var_name='지역', value_name='가격')

# 시간 정보를 숫자로 변환 (인덱스 사용)
df_melted['시간'] = df_melted.index

# 데이터 분할: 학습 데이터와 테스트 데이터 분리
train_data, test_data = train_test_split(df_melted, test_size=0.2, random_state=42)

# 특성과 타겟 분리
X_train = train_data[['지역', '시간']]
y_train = train_data['가격']
X_test = test_data[['지역', '시간']]
y_test = test_data['가격']

# 원-핫 인코더 초기화 및 피팅
encoder = OneHotEncoder(sparse=False)
encoder.fit(df_melted[['지역']])

# 학습 데이터와 테스트 데이터에 원-핫 인코딩 적용
X_train_encoded = encoder.transform(X_train[['지역']])
X_test_encoded = encoder.transform(X_test[['지역']])

# 시간 정보 추가
X_train_encoded = np.hstack([X_train_encoded, X_train[['시간']].to_numpy()])
X_test_encoded = np.hstack([X_test_encoded, X_test[['시간']].to_numpy()])

# RandomForestRegressor 모델 생성 및 학습
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_encoded, y_train)

# 모델 평가
rf_y_pred = rf_model.predict(X_test_encoded)
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_rmse = np.sqrt(rf_mse)

print(rf_rmse)
```

```python
current_time_range = df_melted['시간'].max()
print(f"현재 데이터셋의 최신 시간 인덱스: {current_time_range}")

# 향후 1년 동안의 시간 범위 설정 (예: 현재 데이터셋이 월별 데이터라면, 12개월 추가)
future_time_range = current_time_range + 12

# 각 지역별로 예측 데이터 생성
future_data = []
for 지역 in encoder.categories_[0]:
    for 시간 in range(current_time_range + 1, future_time_range + 1):
        future_data.append([지역, 시간])

# 예측 데이터 프레임 생성
future_df = pd.DataFrame(future_data, columns=['지역', '시간'])

# 원-핫 인코딩 적용
future_encoded = encoder.transform(future_df[['지역']])
future_encoded = np.hstack([future_encoded, future_df[['시간']].to_numpy()])

# 모델을 이용한 예측 수행
predicted_prices = rf_model.predict(future_encoded)

# 예측 결과를 데이터 프레임에 추가
future_df['예측 가격'] = predicted_prices

# 예측 결과 확인
future_df.head()
```
